#### *Exercise*: How do I list all possible categorical values for a particular category?

```python
print(df.INSPECTION_TYPE.unique())
print(df.RESULT.unique())
```

From the openNYC site:
1. INSPECTION_TYPE
    1. Initial Inspection - Inspection conducted in response to a 311 complaint, or a proactive inspection conducted through our neighborhood indexing program.
    2. Compliance Inspection - If a property fails its initial inspection, the Health Department will conduct a follow up (Compliance) inspection.
    3. Baiting - Application of rodenticide, or monitoring visit by a Health Department Pest Control Professional.
    4. Clean Up - The removal of garbage and clutter from a property by the Health Department.
2. RESULTS
    1. Active Rat Signs (ARS) - ARS include any of six different signs: 1) fresh tracks, 2) fresh droppings, 3) active burrows, 4) active runways and rub marks, 5) fresh gnawing marks, and 6) live rats.
    2. Problem Conditions - Problem conditions include garbage (poor containerization of food waste resulting in the feeding of rats), harborage (clutter and dense vegetation promoting the nesting of rats), and mice.

This tells us some interesting info about the potential data points.  Not all points mean rats abound! Lets try to visualize these categories so we can better understand their relationships and distributions. 

```python
df.INSPECTION_TYPE.value_counts().plot(kind='barh', logx=True)
plt.title('Inspection Type')
plt.xlabel('Counts')
plt.figure()
df.RESULT.value_counts().plot(kind='barh', logx=True)
plt.title('Result')
plt.xlabel('Counts')
# Question: Why are we using logx here?
```

These plots tell us something very interesting: Most entries are Initial Inspections, and most Passed Inspection!  So our heatmaps don't really depict a rat-ridden city, but rather a city that's very proactive about inspection and rat-related cleanliness.  It would be nice, however, to see how inspections are broken down by result type.

```python
# This is a useful trick for generating a dataframe that groups two categorical values to visualize the relative breakdown.
# The order goes: GroupBy, Select Top Level Category, Count Aggregation, Unstack Sub Category, Fill in missing values with 0
df_inspect_res = df.groupby(['INSPECTION_TYPE', 'RESULT'])['INSPECTION_TYPE'].count().unstack('RESULT').fillna(0)
```

```python
df_inspect_res
```

```python
df_inspect_res.plot(kind='barh', stacked=True)
plt.title('Inspection and Result')
plt.xlabel('Counts')
```

This shows us what we want... to a degree.  We can't see the values for CLEAN_UPS because there are too few to show up.  Why should/shouldn't we use `logx`?  What else could we do? What else could we improve?

```python
# We need to reshape our dataframe and stick it into a holoviews table to produce the same plot...
s_inpect_res = df_inspect_res.unstack()
s_inpect_res.name='counts'
s_inpect_res
table_inspect_res= hv.Table(s_inpect_res, ['INSPECTION_TYPE', 'RESULT'], 'counts')

table_inspect_res
```

```python
hv.Bars(table_inspect_res, ['INSPECTION_TYPE', 'RESULT'], 'counts').opts(color=hv.Cycle('Category20'), show_legend=True, stacked=True, 
              tools=['hover'], width=800, height=500, invert_axes=True, legend_position='bottom_right')
```

We now have an interactive plot, and we can easily see the breakdown based on inspection and results.  We can use this info and start narrowing down on the rattiest part of NYC.

## Ordinal Data: Ranking, Ordering, and Singling Out

### Univariate Density (Histograms)
Univariate data (in which each datum is a scalar) can be visualized using density distributions.  The most common type of distribution for this case is the histogram, which is an extremely important visualization tool due to its ease of use, interpretability, and commonality in data science.

#### *Exercise*: How can I determine the typical number of reports for a given JOB_ID?
The JOB_PROGRESS category, as we've seen, gives us a great handle for figuring out how many visits a typical job needs.  Lets make a histogram of that feature.

```python
df.JOB_PROGRESS.hist()
plt.title("Distribution of total vists per job? (What's wrong with this?)")
```

A naive attempt to generate a histogram here leads to a bunch of problems.  There's an overwhelming amount of 1s, the binning is extremely course, and the inherent features are completely overshadowed by plotting and analysis errors. First thing's first: we need to make sure we're plotting what we think we're plotting.  If we want the typical total number of visits per job, we need the max JOB_PROGRESS per JOB_ID, not all JOB_PROGRESS.

```python
# For each unique entry (BBL, house, job), get the max value of JOB_PROGRESS
s_max_job_prog = df.groupby(['BBL', 'HOUSE_NUMBER', 'JOB_ID']).JOB_PROGRESS.agg(max)
```

```python
s_max_job_prog.head()
```

Now that the data is cleaned, we still need to tweak the plotting methods to display correctly.

```python
fig, ax = plt.subplots(2,2, figsize=(16,12), sharex=True)
ax = ax.flatten()
s_max_job_prog.hist(ax=ax[0])
ax[0].set_title('Default')
ax[0].set_ylabel('Counts')
s_max_job_prog.hist(ax=ax[1], log=True)
ax[1].set_title('Log Plot')
ax[1].set_xlabel('Total Visits per Job')
s_max_job_prog.hist(ax=ax[2], log=True, bins='auto')
ax[2].set_title('Log Plot with Automatic Binning')
s_max_job_prog.hist(ax=ax[3], log=True, bins=175, range=(1,175))
ax[3].set_title('Log Plot with Fine Binning')
```

Obviously, binning makes a large difference when attempting to display important information in a histogram.  While some amount of smoothing is typical when producing histograms, we are in a special regime in which we are displaying integer values that don't cover a region that is *too* large with respect to the number of points.  Because of this, we can use a very-fine binned histogram (one bin per integer value) to represent our data.  Let's examine our most extreme case and see what's going on there.

```python
# Find the unique entry with the most job followups
max_jobs = s_max_job_prog.max()
print(f"Most appointments for a single job: {max_jobs}")
s_max_job_prog[s_max_job_prog>=max_jobs]
```

### Time Series Data
Before we combine all of our knowledge together, let's take a deeper look at our data with respect to time.  Since we already converted our INSPECTION_DATE to a `datetime` object, displaying our data as a function of time should be fairly easy within the pandas ecosystem.

```python
df.INSPECTION_DATE.hist(bins='auto', log=True)
plt.title('Inspection Date for All Entries')
plt.xlabel('Date (year)')
plt.ylabel('Counts')
```

There are some spurious dates here, as the bulk is obviously between 2010 and mid 2018.  Let's cut out the outliers that may cause us a headache when we try to aggregate events across sections of time.

```python
df_date_trimmed = df[df.INSPECTION_DATE.between('2010-01-01', '2018-06-01')]
```

```python
df_date_trimmed.INSPECTION_DATE.hist(bins='auto', log=True)
plt.title('Inspection Date for Trimmed Entries')
plt.xlabel('Date (year)')
plt.ylabel('Counts')
```

## Combining Knowledge
We've identified key features with respect to this dataset, and can now apply this information to solve our problem.  We should come up with a metric, then plot all locations that fall within that metric.

```python
# Our metric: Total number of "Bait Applied" and "Active Rat Signs" within certain time intervals.

df_date_trimmed['date_cat'] = pd.cut(df_date_trimmed.INSPECTION_DATE, pd.date_range(df_date_trimmed.INSPECTION_DATE.min(), df_date_trimmed.INSPECTION_DATE.max(), freq='3M'))
df_date_trimmed = df_date_trimmed[~df_date_trimmed.date_cat.isnull()]
```

```python
%output holomap='scrubber'
#df_correct_dates = df[df.INSPECTION_DATE>'2015-01-01']
#df_time_range = df[df.INSPECTION_DATE>'2018-01-01']
hvdf = hv.Dataset(df_date_trimmed.query('(RESULT=="Bait applied" or RESULT=="Active Rat Signs") and BORO_CODE==1').sort_values(['date_cat']))
hexplot = hvdf.to(hv.HexTiles, kdims=['X_MERC', 'Y_MERC'], groupby=['date_cat'], dynamic=True)
hexplot.opts(tools=['hover'], colorbar=True) 
hexplot*ny_map
```

```python
df_date_trimmed2 = df_date_trimmed.query('(RESULT=="Bait applied" or RESULT=="Active Rat Signs") and BORO_CODE==1').reset_index()[['ZIP_CODE', 'date_cat', 'RESULT']]

#df_new = pd.DataFrame({'RESULT':df_date_trimmed2.RESULT, })
s_ZIP_date = df_date_trimmed2.groupby(['ZIP_CODE', 'date_cat'])['RESULT'].count()
s_ZIP_date.head()
```

```python
s_ZIP_date = s_ZIP_date.loc[10001:]
```

```python
df_ZIP_date =  s_ZIP_date.unstack(level=0)
df_ZIP_date.head()
```

```python
import hvplot.pandas
```

```python
df_ZIP_date.index = pd.IntervalIndex(df_ZIP_date.index).mid
```

```python
df_ZIP_date.hvplot(legend=False, title='Total Number of Rat Events by ZIP per 3mo', value_label='Rat Events', xlabel='Date', width=900, height=500)
```

```python
df_ZIP_date.cumsum().hvplot(legend=False, title='Cumulative Number of Rat Events by ZIP per 3mo', value_label='Cumulative Rat Events', xlabel='Date', width=900, height=500)
```

### Conclusions
We've found that the majority of adverse rat-events in Manhattan (recorded events in which there was probable signs of rats) occur in the lower east side.  Especially notable is the explosion in rat reports in 2015 for zip code 10003.  However, it seems that there are many addresses associated with 10003 occur outside the expected zone, either due to our misunderstanding of zip-code allotment or due to data-entry error.

Further improvements could be done via normalizing the amount of rat sightings by the total population of the region, by the total area contained in each region, or other metrics.
